```python
import yfinance as yf
```

```python
# Initialize the ticker class, allow us to download price history for a single symbol
sp500 = yf.Ticker("^GSPC")
```

```python
# Display full symbol history from its creation
sp500 = sp500.history(period="max")
```

```python
# Show data in pandas dataframe:
sp500
```

```python
sp500.index
```

```python
# Cleaning and Visualize Data:
# SP500 Price History:
# Show the index (time) on the x-axis and closing price on y-axis
sp500.plot.line(y="Close", use_index=True)
```

```python
# Data cleaning: Remove dividends and stock splits columns:
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
# Setup the target:
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500
```

```python
# Check if tomorrow's price is greater than today's price
# Returns a boolean indicating if tomorrow's price is greater than today's price
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
sp500
```

```python
# Remove all data from before 1990:
sp500 = sp500.loc["2000-01-01":].copy()
```

```python
sp500
```

```python
# Train ML model using Random Forest Classifier (RFC):
# RFC: Trains individual decision trees with randomized parameters, then averages the results from the trees
# RFC is resistent to overfitting, run quickly, pick up non-linear tendencies in data
from sklearn.ensemble import RandomForestClassifier

# n = number of decision trees we want to train, higher value = better accuracy
model = RandomForestClassifier(n_estimators=150, min_samples_split=100, random_state=1)

# Typical data set split: 80% train, 20% test
# Train set: All rows except the last 1254 
# Test set: Last 1254 rows
train = sp500.iloc[:-1254]
test = sp500.iloc[-1254:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
# Measure model accuracy:
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
import pandas as pd
preds = pd.Series(preds, index=test.index)
```

```python
precision = precision_score(test["Target"], preds, zero_division=0)
```

```python
# Shows what percentage of the whole time the stock went up:
precision
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
# Orange is predictions, blue is what actually happened
combined.plot()
```

```python
# More robust way to test outcome:

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
# 0 = went down that many days
# 1 = went up that many days
predictions["Predictions"].value_counts()
```

```python
# Market went up this precent of the time:
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
# Percentage of days the market went up:
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
# Add more predictors to improve accuracy:
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_column, trend_column]
```

```python
sp500
```

```python
sp500 = sp500.dropna()
```

```python
sp500
```

```python
# Improving the model:
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
# New precision score with new predictors and thresholds set
# When the model predicts the stock will go up, it does go up about 54% of the time
precision_score(predictions["Target"], predictions["Predictions"])
```
